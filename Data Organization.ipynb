```python
import pandas as pd

# Assuming your dataframes are loaded as follows for Miami:
# Miami_daily_dataframe_sc_new, Miami_daily_dataframe_noaa_new, Miami_daily_dataframe_vc_new, Miami_daily_dataframe_om_new

# Merge the first two dataframes on 'Date' with an outer join
Miami_merged = pd.merge(Miami_daily_dataframe_sc_new, Miami_daily_dataframe_noaa_new, on='Date', how='outer', suffixes=('_sc', '_noaa'))
Miami_merged = pd.merge(Miami_merged, Miami_daily_dataframe_vc_new, on='Date', how='outer', suffixes=('', '_vc'))
Miami_merged = pd.merge(Miami_merged, Miami_daily_dataframe_mt_new, on='Date', how='outer', suffixes=('', '_mt'))
Miami_merged = pd.merge(Miami_merged, Miami_daily_dataframe_om_new, on='Date', how='outer', suffixes=('', '_om'))
import pandas as pd
import numpy as np


temperature_columns = ['maxtemp_sc', 'maxtemp_noaa', 'maxtemp_vc', 'maxtemp_mt', 'maxtemp_om']

# Calculate the average of the temperature columns and assign it to a new column 'avg_max_temp'
Miami_merged['max_temp'] = Miami_merged[temperature_columns].mean(axis=1).round()

# If there are prefix or suffix issues in the column names due to merging, ensure the temperature_columns list is updated to match the final merged dataframe's column names exactly.


# Miami_mergednow contains combined data for Miami from all sources, with NaNs allowed for missing data
Miami_merged
```

```python
# Step 1: Merge the Austin dataframes
Austin_merged = pd.merge(Austin_daily_dataframe_sc_new, Austin_daily_dataframe_noaa_new, on='Date', how='outer', suffixes=('_sc', '_noaa'))
Austin_merged = pd.merge(Austin_merged, Austin_daily_dataframe_vc_new, on='Date', how='outer')
Austin_merged = pd.merge(Austin_merged, Austin_daily_dataframe_mt_new, on='Date', how='outer')
Austin_merged = pd.merge(Austin_merged, Austin_daily_dataframe_om_new, on='Date', how='outer')

# Step 2: Calculate the average max temp
temperature_columns_austin = ['maxtemp_sc', 'maxtemp_noaa', 'maxtemp_vc', 'maxtemp_mt', 'maxtemp_om']
Austin_merged['max_temp'] = Austin_merged[temperature_columns_austin].mean(axis=1).round()

# Austin_merged now contains combined data for Austin from all sources, with handling for NaNs
Austin_merged
```

```python
# Merging Central Park dataframes
Central_Park_merged = pd.merge(Central_park_daily_dataframe_sc_new, Central_park_daily_dataframe_noaa_new, on='Date', how='outer', suffixes=('_sc', '_noaa'))
Central_Park_merged = pd.merge(Central_Park_merged, Central_park_daily_dataframe_vc_new, on='Date', how='outer')
Central_Park_merged = pd.merge(Central_Park_merged, Central_park_daily_dataframe_mt_new, on='Date', how='outer')
Central_Park_merged = pd.merge(Central_Park_merged, Central_park_daily_dataframe_om_new, on='Date', how='outer')

# Calculate the average max temp for Central Park
temperature_columns_cp = ['maxtemp_sc', 'maxtemp_noaa', 'maxtemp_vc', 'maxtemp_mt', 'maxtemp_om']
Central_Park_merged['max_temp'] = Central_Park_merged[temperature_columns_cp].mean(axis=1).round()
Central_Park_merged
```

```python
# Merging Chicago dataframes
Chicago_merged = pd.merge(Chicago_daily_dataframe_sc_new, Chicago_daily_dataframe_noaa_new, on='Date', how='outer', suffixes=('_sc', '_noaa'))
Chicago_merged = pd.merge(Chicago_merged, Chicago_daily_dataframe_vc_new, on='Date', how='outer')
Chicago_merged = pd.merge(Chicago_merged, Chicago_daily_dataframe_mt_new, on='Date', how='outer')
Chicago_merged = pd.merge(Chicago_merged, Chicago_daily_dataframe_om_new, on='Date', how='outer')

# Calculate the average max temp for Chicago
temperature_columns_chicago = ['maxtemp_sc', 'maxtemp_noaa', 'maxtemp_vc', 'maxtemp_mt', 'maxtemp_om']
Chicago_merged['max_temp'] = Chicago_merged[temperature_columns_chicago].mean(axis=1).round()
Chicago_merged
```

```python
# Filter out rows from 2024-01-01 and later for each city's dataframe
Miami_merged = Miami_merged[Miami_merged['Date'] < '2024-01-01']
Austin_merged = Austin_merged[Austin_merged['Date'] < '2024-01-01']
Central_Park_merged = Central_Park_merged[Central_Park_merged['Date'] < '2024-01-01']
Chicago_merged = Chicago_merged[Chicago_merged['Date'] < '2024-01-01']

# Proceed with adding the 'City' column, concatenating, and saving as before
Miami_merged['City'] = 'Miami'
Austin_merged['City'] = 'Austin'
Central_Park_merged['City'] = 'Central Park'
Chicago_merged['City'] = 'Chicago'

# Concatenate all city dataframes into one, ensuring to include the 'City' column
all_cities_merged = pd.concat([Miami_merged, Austin_merged, Central_Park_merged, Chicago_merged], ignore_index=True)

# Assuming the 'Date' column is of string type. If it's datetime, ensure it's converted before the comparison or adjust the comparison accordingly

# Reorder the dataframe columns to include the 'City' column
columns_order = ['Date', 'maxtemp_sc', 'maxtemp_noaa', 'maxtemp_vc', 'maxtemp_mt', 'maxtemp_om', 'max_temp', 'City']
all_cities_merged = all_cities_merged[columns_order]

# Save to an Excel file
excel_file_path = 'data/all_cities_weather_filtered.xlsx'  # Adjust path as needed
all_cities_merged.to_excel(excel_file_path, index=False)

excel_file_path
```
